```python
import numpy as np
import matplotlib.pyplot as plt

# 1) Generar y graficar 1 caminata aleatoria
rng = np.random.default_rng(0)
n_steps = 500
steps = rng.choice([-1, 1], size=n_steps)
walk = np.cumsum(steps)

plt.figure(figsize=(8, 4))
plt.plot(walk, lw=1.5)
plt.title('Caminata aleatoria (1 trayectoria)')
plt.xlabel('Paso')
plt.ylabel('Posicion')
plt.grid(alpha=0.3)
plt.show()
```

```python
# 2) Distribucion final de N caminatas aleatorias
rng = np.random.default_rng(1)
n_steps = 300
n_walks = 10000
steps = rng.choice([-1, 1], size=(n_walks, n_steps))
final_positions = steps.sum(axis=1)

plt.figure(figsize=(7, 4))
plt.hist(final_positions, bins=40, edgecolor='white')
plt.title('Distribucion final de N caminatas aleatorias')
plt.xlabel('Posicion final')
plt.ylabel('Frecuencia')
plt.grid(alpha=0.2)
plt.show()
```

```python
# 3) Varianza de resultados para diferentes largos de caminatas
rng = np.random.default_rng(2)
lengths = np.array([50, 100, 200, 400, 800, 1200])
n_walks = 5000
variances = []

for n_steps in lengths:
    steps = rng.choice([-1, 1], size=(n_walks, n_steps))
    final_positions = steps.sum(axis=1)
    variances.append(final_positions.var(ddof=0))

plt.figure(figsize=(7, 4))
plt.plot(lengths, variances, marker='o')
plt.title('Varianza vs largo de la caminata')
plt.xlabel('Numero de pasos')
plt.ylabel('Varianza de la posicion final')
plt.grid(alpha=0.3)
plt.show()
```

```python
# 4) Aproximacion de Movimiento Browniano con random walk
rng = np.random.default_rng(3)
T = 1.0
n_steps = 1000
dt = T / n_steps
steps = rng.choice([-1, 1], size=n_steps) * np.sqrt(dt)
brownian = np.cumsum(steps)
t = np.linspace(0, T, n_steps)

plt.figure(figsize=(8, 4))
plt.plot(t, brownian, lw=1.5)
plt.title('Aproximacion de Movimiento Browniano (random walk)')
plt.xlabel('t')
plt.ylabel('B(t)')
plt.grid(alpha=0.3)
plt.show()
```

```python
# 5) 200 caminatas aleatorias y su promedio en cada t
rng = np.random.default_rng(4)
n_steps = 300
n_walks = 200
steps = rng.choice([-1, 1], size=(n_walks, n_steps))
walks = np.cumsum(steps, axis=1)
mean_walk = walks.mean(axis=0)

plt.figure(figsize=(9, 5))
plt.plot(walks.T, color='steelblue', alpha=0.15, lw=1)
plt.plot(mean_walk, color='black', lw=2, label='Promedio')
plt.title('200 caminatas aleatorias y promedio')
plt.xlabel('Paso')
plt.ylabel('Posicion')
plt.legend()
plt.grid(alpha=0.2)
plt.show()
```
